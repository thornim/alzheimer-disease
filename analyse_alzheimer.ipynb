### 5. Analyse des dépendances

__Test de normalité__

```python
# Sélection de toutes les variables à tester :
vars = ["MemoryComplaints", "Hypertension", "Diabetes", "Depression", "Diagnosis", "PhysicalActivity", "MMSE", "FunctionalAssessment", "ADL"]

df_test = df[vars].copy()

# Vérif :
df_test.head()
```

- Shapiro test :

```python
# Test Shapiro uniquement sur les variables continues : 
import scipy.stats as st

PA = df_test["PhysicalActivity"]
MSE = df_test["MMSE"]
FA = df_test["FunctionalAssessment"]
ADL = df_test["ADL"]
```

```python
# Résultats sous forme de résumé dans un DataFrame :
import pandas as pd
import scipy.stats as st

variables = {
    "PhysicalActivity": PA,
    "MMSE": MSE,
    "FunctionalAssessment": FA,
    "ADL": ADL
}

rows = []
alpha = 0.05

for name, series in variables.items():
    stat, p = st.shapiro(series)
    rows.append({
        "Variable": name,
        "Statistic": stat,
        "p_value": p,
        "Normale (p > 0.05)": p > alpha
    })

shapiro_df = pd.DataFrame(rows)
shapiro_df
```

- QQ-plot : 

```python
import statsmodels.api as sm 
fig = sm.qqplot(PA, fit=True, line='45')
```

```python
import statsmodels.api as sm 
fig = sm.qqplot(MSE, fit=True, line='45')
```

```python
import statsmodels.api as sm 
fig = sm.qqplot(FA, fit=True, line='45')
```

```python
import statsmodels.api as sm 
fig = sm.qqplot(ADL, fit=True, line='45')
```

- Hypothèse nulle : la variable suit une loi normale.
- Hypothèse alternative : la variable ne suit pas une loi normale.

Si p <= 0.05 alors absence de normalité.
Si p > 0.05 on peut dire qu'il y a une normalité "apparente".

Dans notre étude, on considérera que les variables MMSE et FA (FunctionalAssessment) sont normales.

### Lien entre une variable numérique et une variable catégorielle

Nous allons comparer la distribution d’une variable quantitative (PA, MMSE, FA, ADL) entre k groupes définis par la variable catégorielle (Diagnostic).

__Test T Student (2 classes)__

```python
import pandas as pd
from scipy import stats

tests_t = []

for var in ["MMSE", "FunctionalAssessment"]:
    g0 = df_test.loc[df["Diagnosis"] == 0, var]
    g1 = df_test.loc[df["Diagnosis"] == 1, var]
    t_stat, p_val = stats.ttest_ind(g0, g1, equal_var=False)
    tests_t.append({
        "Variable": var,
        "Moyenne Diagnosis=0": g0.mean(),
        "Moyenne Diagnosis=1": g1.mean(),
        "t_stat": t_stat,
        "p_value": p_val,
        "Différence significative (p <= 0.05)": p_val <= 0.05
    })

t_df = pd.DataFrame(tests_t)
t_df
```

__Test de Kruskal-Wallis (non-paramétrique)__

```python
from scipy import stats

# Séparation des groupes
pa_0  = df_test.loc[df["Diagnosis"] == 0, "PhysicalActivity"]
pa_1  = df_test.loc[df["Diagnosis"] == 1, "PhysicalActivity"]

adl_0 = df_test.loc[df["Diagnosis"] == 0, "ADL"]
adl_1 = df_test.loc[df["Diagnosis"] == 1, "ADL"]

# Kruskal-Wallis 
kw_pa_stat, kw_pa_p = stats.kruskal(pa_0, pa_1)
kw_adl_stat, kw_adl_p = stats.kruskal(adl_0, adl_1)

print(f"PhysicalActivity ~ Diagnosis : H = {kw_pa_stat:.3f}, p = {kw_pa_p:.4f}")
print(f"ADL ~ Diagnosis              : H = {kw_adl_stat:.3f}, p = {kw_adl_p:.4f}")
```

### Lien entre deux variables catégorielles

__Test du Khi-2__

```python
import pandas as pd
from scipy import stats

observed = pd.crosstab(df_test["Hypertension"], df_test["Diagnosis"])
chi2_stat, p_value, dof, expected = stats.chi2_contingency(observed)

print(f"Chi² = {chi2_stat:.3f}, ddl = {dof}, p = {p_value:.4f}")
expected = pd.DataFrame(expected, index=observed.index, columns=observed.columns)
expected
```
